```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load data
df = pd.read_csv('all_logs.csv')

# Convert temperature to integer
df['Temperature'] = df['Temperature'].astype(int)

# Convert node to string
df['Node'] = df['Node'].astype(str)

# Convert GPU to string
df['GPU'] = df['GPU'].astype(str)

# Create a new column combining Node and GPU
df['Node_GPU'] = df['Node'] + '_' + df['GPU']

# Create the violinplot
plt.figure(figsize=(16, 9))
sns.violinplot(x='Node_GPU', y='Temperature', data=df, scale='width', inner='quartile')

# Add title and labels
plt.title('GPU Temperatures over Time', fontsize=20)
plt.xlabel('Node_GPU', fontsize=16)
plt.ylabel('Temperature (C)', fontsize=16)

# Rotate x labels for better readability
plt.xticks(rotation=90)

# Show plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load data
df = pd.read_csv('all_logs.csv')

# Convert temperature to integer
df['Temperature'] = df['Temperature'].astype(int)

# Convert node to string
df['Node'] = df['Node'].astype(str)

# Convert GPU to string
df['GPU'] = df['GPU'].astype(str)

# Get a list of unique nodes
nodes = df['Node'].unique()

# Create subplots for each node
fig, axs = plt.subplots(len(nodes), 1, figsize=(10, 5*len(nodes)))

for i, node in enumerate(nodes):
    # Filter data for the current node
    df_node = df[df['Node'] == node]
    
    # Create violin plot
    sns.violinplot(x='GPU', y='Temperature', data=df_node, scale='width', inner='quartile', ax=axs[i])
    
    # Set title and labels
    axs[i].set_title('Node ' + node, fontsize=20)
    axs[i].set_xlabel('GPU', fontsize=16)
    axs[i].set_ylabel('Temperature (C)', fontsize=16)
    
    # Rotate x labels for better readability
    axs[i].set_xticklabels(axs[i].get_xticks(), rotation=90)

# Adjust layout to prevent overlap of subplots
plt.tight_layout()

# Show plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load data
df1 = pd.read_csv('all_logs.csv')
df2 = pd.read_csv('node207_logs.csv')

# Add an extra column for machine label
df1['Machine'] = '3080/3090'
df2['Machine'] = '4090'

# Concatenate the two dataframes
df = pd.concat([df1, df2])

# Convert temperature and utilization to integer
df['Temperature'] = df['Temperature'].astype(int)
df['Utilization'] = df['Utilization'].astype(int)

# Convert Node to string
df['Node'] = df['Node'].astype(str)

fig, axs = plt.subplots(1, 2, figsize=(15, 6))

# Create box plot for temperature
sns.boxplot(x='Node', y='Temperature', hue='Machine', data=df, ax=axs[0])
axs[0].set_title('Temperature Distribution by Node', fontsize=20)
axs[0].set_xlabel('Node', fontsize=16)
axs[0].set_ylabel('Temperature (C)', fontsize=16)
axs[0].set_xticklabels(axs[0].get_xticklabels(), rotation=45)

# Create box plot for utilization
sns.boxplot(x='Node', y='Utilization', hue='Machine', data=df, ax=axs[1])
axs[1].set_title('Utilization Distribution by Node', fontsize=20)
axs[1].set_xlabel('Node', fontsize=16)
axs[1].set_ylabel('Utilization (%)', fontsize=16)
axs[1].set_xticklabels(axs[1].get_xticklabels(), rotation=45)
axs[1].legend().remove()  # Remove legend for the second subplot

# Show plots
plt.tight_layout()
plt.show()
```
